```python
import pandas as pd 

#Read the data
pumpkins = pd.read_csv('US-pumpkins.csv')

#Print the first 5 rows
pumpkins.head()
```

```python
 #Filter the data
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
pumpkins.head()
```

```python
#Count how many null values are in the data
pumpkins.isnull().sum()
```

```python
#Keep only the columns we need
columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date']

#Drop all other columns
pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in columns], axis=1)
```

```python
  #Calculate the avarage price
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2
```

```python
#Get the month from the date
month = pd.DatetimeIndex(pumpkins['Date']).month
```

```python
#Create a new dataframe

new_pumpkins = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'], 'High Price': pumpkins['High Price'], 'Price': price})
new_pumpkins.head()
```

```python
#Conver package to numeric
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price / 1.1
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price / 0.5
```

```python
new_pumpkins.tail()
```

```python
import matplotlib.pyplot as plt

#Plot the data
price = new_pumpkins['Price']
month = new_pumpkins['Month']

#show a scatter plot of the data
plt.scatter(price, month)
plt.show()
```

```python
new_pumpkins.groupby(['Month'])['Price'].mean().plot(kind='bar')

#Show the plot and labal the axes
plt.ylabel('Price')
```
